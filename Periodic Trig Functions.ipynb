```python
import numpy as np
from matplotlib import pyplot as plt
```

### Plot Sin and Cos 

```python
def create_cartesian_plane(x=4, y=1):
    ax, fig = plt.subplots()
    plt.axvline(0, color='k', linewidth=0.5)
    plt.axhline(0, color='k', linewidth=0.5)
    plt.xlim(-1*x, x)
    plt.ylim(-1*y, y)
    
x_coords  = np.arange(-4, 4, 0.01)
sine = np.sin(x_coords)
cosine = np.cos(x_coords)

create_cartesian_plane(x=4)
plt.plot(x_coords,sine, color='indigo')
plt.plot(x_coords, cosine, color='violet')
plt.show()
```
